```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.array([5, 15, 25, 35, 45, 55]).reshape(-1, 1)  # Reshape x to be a 2D array
y = np.array([5, 20, 14, 32, 22, 38])  # y can remain 1D, scikit-learn handles it

print(x)
print(y)

print(x.shape)
print(y.shape)

```

```python
model = LinearRegression()
model.fit(x, y)  # Now the fit method should work without errors
```

```python
# to check whether the model works satisfactorily,
# obtain the coefficient of determination, R2, with .score()
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
```

```python
# The attributes of model are .intercept_ (b0), and .coef_ (b1)
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")
```

```python
# Instead of predict(), You can also use below equation for test reponses.
y_pred = model.intercept_ + model.coef_ * x
print(f"predicted response:\n{y_pred}")
```

```python
plt.scatter(x,y, color = 'red')
plt.plot(x, y_pred, color = 'blue')
plt.title('y vs x (Training set)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
