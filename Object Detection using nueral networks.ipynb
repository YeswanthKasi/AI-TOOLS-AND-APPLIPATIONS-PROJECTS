```python
import cv2
import numpy as np

# Load the YOLOv3 model
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")

# Load the COCO class labels
classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Define the detection function
def detect_objects(image):
    # Create a blob from the image
    blob = cv2.dnn.blobFromImage(image, 1/255, (416, 416), swapRB=True, crop=False)

    # Set the input for the network
    net.setInput(blob)

    # Run the forward pass
    outputs = net.forward(net.getUnconnectedOutLayersNames())

    # Initialize the detected objects
    detected_objects = []

    # Iterate over the detections
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5 and class_id == 0:  # Filter out low-confidence detections and non-person objects
                x, y, w, h = detection[0:4] * np.array([image.shape[1], image.shape[0], image.shape[1], image.shape[0]])
                detected_objects.append((x, y, w, h, class_id))

    return detected_objects

# Load an example image
image = cv2.imread("example_image.jpg")

# Detect objects in the image
detected_objects = detect_objects(image)

# Draw bounding boxes around the detected objects
for obj in detected_objects:
    x, y, w, h, class_id = obj
    cv2.rectangle(image, (int(x), int(y)), (int(x+w), int(y+h)), (0, 255, 0), 2)
    cv2.putText(image, classes[class_id], (int(x), int(y-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Display the output
cv2.imshow("Object Detection", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
